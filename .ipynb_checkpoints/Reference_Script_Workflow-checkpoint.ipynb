```python
# importation des librairies
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
np.random.seed(1)
```

```python
# lecture du fichier et randomisation des données
csv_listing = pd.read_csv('paris_airbnb.csv')
csv_listing = csv_listing.loc[np.random.permutation(len(csv_listing))]
csv_listing
```

```python
# nettoyage des données pour retirer les caractères non numérique
clean_virgule = csv_listing['price'].str.replace(',','')
clean_dollars = clean_virgule.str.replace('$','')
csv_listing['price'] = clean_dollars.astype('float')
csv_listing
```

```python
# visualisation des infos du fichier CSV
csv_listing.info()
```

```python
# nettoyage des colonnes qui ne serve pas au calcule du dataSet
drop_columns = ['cleaning_fee','security_deposit','city','zipcode','state','room_type','latitude','longitude','host_response_rate','host_acceptance_rate','host_listings_count']
csv_listing = csv_listing.drop(drop_columns, axis=1)
```

```python
# suppression des données null dans les diverses colonnes
csv_listing = csv_listing.dropna(axis=0)
print(csv_listing.isnull().sum())
```

```python
# normalisation des données pour facilité les calcules
normalized_listing = (csv_listing - csv_listing.mean()) / csv_listing.std()
print(normalized_listing)
```

```python
# creation de la variable KNN
knn = KNeighborsRegressor(n_neighbors=5,algorithm ='brute')
```

```python
# division du dataSet pour entrainer le model
train_df = normalized_listing.iloc[0:6000]
test_df = normalized_listing.iloc[6000:]
# variable pour définir les colonnes qui vont être comparer
train_columns = ['accommodates','bedrooms']
# utilisation de la fonction fit
knn.fit(train_df[train_columns], train_df['price'])
# utilisation de la fonction predict
prediction = knn.predict(test_df[train_columns])
prediction
```
